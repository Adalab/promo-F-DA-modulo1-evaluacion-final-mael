inventario (lista de diccionarios): Un atributo para almacenar los productos en el inventario. Cada producto debe ser representado como un diccionario con las siguientes claves: 'nombre', 'precio', y 'cantidad'. Al principio deberá ser una lista vacía. Ejemplo de como debería ser:

```python
import re

class Tiendaonline:
    def __init__(self, inventario, clientes):
        self.inventario = inventario
        self.clientes = clientes
        self.ventas_totales = 0.0


    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto['nombre'].lower() == nombre.lower():
                producto['cantidad'] += cantidad
                return
        nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
        self.inventario.append(nuevo_producto)

    def ver_inventario(self):
        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")


    def manejar_accion(self):
        accion = input("Ingrese 'buscar' para encontrar un producto o 'actualizar' para cambiar el stock: ").lower()
        if accion == 'buscar':
            self.buscar_producto()
        elif accion == 'actualizar':
            nombre = input("Ingrese el nombre del producto: ").lower()
            self.actualizar_stock(nombre)
        else:
            print("Acción no reconocida. Inténtelo de nuevo.")

    def buscar_producto(self):
        nombre_buscado = input("Introduce el nombre del producto a buscar: ").lower()
        for producto in self.inventario:
            if producto['nombre'].lower() == nombre_buscado:
                print(f"Nombre: {producto['nombre'].lower()}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return
        print("El producto no está en el inventario")

    def actualizar_stock(self, nombre):
        encontrado = False
        for producto in self.inventario:
            if producto['nombre'].lower() == nombre:
                accion = input("Ingrese 'agregar' para sumar cantidad o 'quitar' para restar cantidad: ").lower()
                cantidad = int(input("Ingrese la cantidad: "))
                if accion == 'agregar':
                    producto['cantidad'] += cantidad
                    encontrado = True
                    break
                elif accion == 'quitar':
                    producto['cantidad'] -= cantidad
                    encontrado = True
                    break
                else:
                    print("Acción no válida. Intente de nuevo.")
                    encontrado = True
                    break
        if not encontrado:
            print("\nEl producto no está en el inventario")

    def calcular_valor_inventario(self):
        valor_total = 0
        for producto in self.inventario:
            valor_total += producto['precio'] * producto['cantidad']
        print(f"\nEl valor total de mi inventario es: {valor_total} euros")

    def buscar_producto_regex(self):
        patron_busqueda = input("Introduce el patrón de búsqueda en formato regex: ").lower()
        resultados = []
        for producto in self.inventario:
            if re.findall(patron_busqueda, producto['nombre'].lower()):
                resultados.append(producto)
        if resultados:
            print("\nResultados de la búsqueda con regex:")
            for producto in resultados:
                print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
        else:
            print("No se encontraron coincidencias.")


    def realizar_compra(self):
        carrito = []
        costo_total = 0

        while True:
            print("\nInventario disponible:")
            self.ver_inventario()
            nombre_producto = input("\nIngrese el nombre del producto que desea comprar: ").lower()

            encontrado = False
            for producto in self.inventario:
                if producto['nombre'].lower() == nombre_producto and producto['cantidad'] > 0:
                    carrito.append(producto)
                    costo_total += producto['precio']
                    producto['cantidad'] -= 1
                    encontrado = True
                    break

            if not encontrado:
                print("El producto no está disponible en el inventario.")

            continuar = input("\n¿Desea comprar algo más? (si/no): ").lower()
            if continuar == 'no':
                break

        print("\nProductos en su carrito:")
        for producto in carrito:
            print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}")
        print(f"\nEl costo total de su compra es: ${costo_total}")

        confirmacion_pago = input("¿Desea pagar? (si/no): ")
        if confirmacion_pago.lower() == 'si':
            cantidad_pagada = float(input("Ingrese la cantidad con la que va a pagar: $"))
            if cantidad_pagada < costo_total:
                print("La cantidad ingresada es insuficiente.")
            else:
                cambio = cantidad_pagada - costo_total
                print(f"Su cambio es: ${cambio}")
                self.registrar_compra(costo_total, nombre_producto)
        else:
            print("Compra cancelada.")

    def registrar_compra(self, costo_total, nombre_producto):
        nombre_cliente = input("Ingrese el nombre del cliente: ")
        cliente_encontrado = False
        for cliente in self.clientes:
            if cliente['nombre'] == nombre_cliente:
                if 'total_gastado' not in cliente:
                    cliente['total_gastado'] = costo_total
                    cliente['compras'] = [{"producto": nombre_producto, "precio": costo_total, "cantidad": 1, "total": costo_total}]
                else:
                    cliente['total_gastado'] += costo_total
                    cliente['compras'].append({"producto": nombre_producto, "precio": costo_total, "cantidad": 1, "total": costo_total})
                self.ventas_totales += costo_total
                cliente_encontrado = True
                print("Compra registrada con éxito.")
                self.agregar_cliente(nuevo_nombre, nuevo_email, nombre_producto)
                break

        if not cliente_encontrado:
            print("Cliente no encontrado.")
            confirmacion_agregar = input("¿Desea agregar a este cliente? (si/no): ")
            if confirmacion_agregar.lower() == 'si':
                nuevo_nombre = nombre_cliente
                nuevo_email = input("Ingrese el email del nuevo cliente: ")
                self.agregar_cliente(nuevo_nombre, nuevo_email, nombre_producto)


                
    def agregar_cliente(self, nombre, email, nombre_producto):
        total_compra = 0
        for producto in self.inventario:
            if producto['nombre'] == nombre_producto:
                total_compra = producto['precio']
                break
        compras_cliente = [{"producto": nombre_producto, "precio": total_compra, "cantidad": 1, "total": total_compra}]
        cliente = {'nombre': nombre, 'email': email, 'compras': compras_cliente, "Total": total_compra}
        self.clientes.append(cliente)
    
    def ver_clientes(self):
        for cliente in self.clientes:
            if 'compras' not in cliente:
                cliente['compras'] = []
            print(f"\nNombre: {cliente['nombre']}, Email: {cliente['email']}, Compras: {cliente['compras']}, Total: {cliente['Total']}")
            
    def ver_compras_cliente(self, nombre_cliente):
        for cliente in self.clientes:
            if cliente['nombre'] == nombre_cliente:
                if 'compras' in cliente:
                    print(f"\nHistorial de compras de {nombre_cliente}:")
                    for compra in cliente['compras']:
                        print(f"Producto: {compra['producto']}, Precio: {compra['precio']}, Cantidad: {compra['cantidad']}, Total: {compra['total']}")
                else:
                    print(f"\nEl cliente {nombre_cliente} no tiene historial de compras.")
                return
        print(f"\nNo se encontró al cliente {nombre_cliente}.")
        
    def calcular_ventas_totales(self):
        ventas_totales_tienda = 0
        for cliente in self.clientes:
            if 'compras' in cliente:
                for compra in cliente['compras']:
                    ventas_totales_tienda += compra['total']
        print(f"\nLas ventas totales de la tienda son: {ventas_totales_tienda}")
           

inventario = [
    {"nombre": "Camisa", "precio": 20, "cantidad": 40},
    {"nombre": "Pantalón", "precio": 30, "cantidad": 30},
]

clientes = [
    {"nombre": "Maria Gonzalez", "email": "cliente1@email.com", "compras": [{"producto": "Bolso", "precio": 15, "cantidad": 1, "total": 15}], "Total": 15},
    {"nombre": "Juan Perez", "email": "cliente2@email.com", "compras" : [{"producto": "Pantalón", "precio": 30, "cantidad": 5, "total": 150}], "Total": 150}
]

tienda = Tiendaonline(inventario, clientes)
print("Inventario al iniciar el día")
tienda.ver_inventario()


tienda.agregar_producto("Bolso", 15, 25)
tienda.agregar_producto("Zapatillas", 70, 10)
tienda.agregar_producto("Botas", 90, 15)
tienda.agregar_producto("Calcetines", 10, 100)
tienda.agregar_producto("Pantalón", 30, 100)

#print("\nEste es mi inventario después de agregar productos")
#tienda.ver_inventario()


#tienda.manejar_accion()

#print("\nEl producto buscado:")
#tienda.buscar_producto()

#tienda.buscar_producto_regex()

#tienda.realizar_compra()

#nuevo_nombre = input("Ingrese el nombre del nuevo cliente: ")
#nuevo_email = input("Ingrese el email del nuevo cliente: ")
#tienda.agregar_cliente(nuevo_nombre, nuevo_email)


print("Estos son mis clientes:")
tienda.ver_clientes()

nombre_cliente = input("Ingrese el nombre del cliente para ver su historial de compras: ")
tienda.ver_compras_cliente(nombre_cliente)

tienda.calcular_ventas_totales()

tienda.calcular_valor_inventario()
```
